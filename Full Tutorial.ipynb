### Craigslist URL

```python
def find_posting_url(li_soup):
    return li_soup.find_all('a', class_='result-title hdrlnk')[0]['href']
```

```python
for li_soup in soup.find_all('li', class_= 'result-row'):
    print(find_posting_url(li_soup))
```

### Craigslist Date Posted

```python
def find_date_posted(li_soup):
    return datetime.strptime(str(li_soup.select('.result-date')[0]).split('datetime="')[1][:16], '%Y-%m-%d %H:%M')
```

```python
def find_date_posted2(li_soup):  #BETTER FORM!!
    return datetime.strptime(li_soup.select('.result-date')[0]['datetime'], '%Y-%m-%d %H:%M')
```

```python
find_date_posted2(soup.find_all('li', class_= 'result-row')[0])
```

```python
for li_soup in soup.find_all('li', class_= 'result-row'):
    print(find_date_posted2(li_soup))
```

```python
for li_soup in soup.find_all('li', class_= 'result-row'):
    print(find_date_posted(li_soup))
```

### Craigslist Title

```python
def find_post_title(li_soup):
    return li_soup.find_all('a', class_='result-title hdrlnk')[0].getText()
```

```python
for li_soup in soup.find_all('li', class_= 'result-row'):
    print(find_post_title(li_soup))
```

### Craigslist Price

```python
def find_post_price(li_soup):
    return int(li_soup.find_all('span', class_='result-price')[0].getText().strip('$').replace(',',''))
```

```python
for li_soup in soup.find_all('li', class_= 'result-row'):
    print(find_post_price(li_soup))
```

### Craigslist Bedrooms

```python
def find_num_bdrms(li_soup):
    return int([x for x in li_soup.find_all('span', class_='housing')[0].getText() if x.isdigit()][0])
```

```python
for li_soup in soup.find_all('li', class_= 'result-row'):
    print(find_num_bdrms(li_soup))
```

## Using XPATH
uses lxml.html, requests


```python
tree = lxml.html.fromstring(ycomb_response.text)
```

### Simplest Approach

```python
title_xpath = '//*[contains(concat( " ", @class, " " ), concat( " ", "storylink", " " ))]'
source_xpath = '//*[contains(concat( " ", @class, " " ), concat( " ", "sitestr", " " ))]'
points_xpath = '//*[contains(concat( " ", @class, " " ), concat( " ", "score", " " ))]'
author_path = '//*[contains(concat( " ", @class, " " ), concat( " ", "hnuser", " " ))]'
post_age_path = '//*[contains(concat( " ", @class, " " ), concat( " ", "age", " " ))]//a'
num_comments_path = '//*[contains(concat( " ", @class, " " ), concat( " ", "subtext", " " ))]//a+//a'
```

```python
title = [x.text_content() for x in tree.xpath(title_xpath)]
source = [x.text_content() for x in tree.xpath(source_xpath)]
points = [x.text_content() for x in tree.xpath(points_xpath)]
author = [x.text_content() for x in tree.xpath(author_path)]
post_age = [x.text_content() for x in tree.xpath(post_age_path)]
# num_comments = [x.text_content() for x in tree.xpath(num_comments_path)]
```
